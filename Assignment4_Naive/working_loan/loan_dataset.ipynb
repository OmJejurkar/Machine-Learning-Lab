```python
import pandas as pd
df = pd.read_csv('loan_data.csv')
df.head()
```

```python
df.info()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.countplot(data=df,x='purpose',hue='not.fully.paid')
plt.xticks(rotation=45,ha = 'right')
```

```python
pref_df = pd.get_dummies(df,columns=['purpose'],drop_first=True)
pref_df.head()
```

```python
from sklearn.model_selection import train_test_split
X = pref_df.drop('not.fully.paid',axis=1)
Y = pref_df['not.fully.paid']
X_train, X_test, Y_train,Y_test = train_test_split(X,Y,test_size=0.3,random_state=101)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(X_train,Y_train)
```

```python
from sklearn.metrics import (
    accuracy_score,
    confusion_matrix,
    ConfusionMatrixDisplay,
    f1_score,
    classification_report,
)

y_pred = model.predict(X_test)

accuray = accuracy_score(Y_test, y_pred)
f1 = f1_score(Y_test, y_pred, average="weighted")

print("Accuracy:", accuray)
print("F1 Score:", f1)
```

```python
labels = ["not.Fully Paid", "Not fully Paid"]
cm = confusion_matrix(Y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=labels)
disp.plot()
```

```python
from sklearn.metrics import classification_report
target_names = ['loan', 'not loan']
print(classification_report(Y_test, y_pred, target_names=target_names))
```

```python
#calculate specificity
tn, fp, fn, tp = cm.ravel()
print("True Negatives:", tn)
print("False Positives:", fp)
print("False Negatives:", fn)
print("True Positives:", tp)
print("Specificity:", tn / (tn + fp))  # Specificity = TN / (TN + FP)
print("Sensitivity:", tp / (tp + fn))  # Sensitivity = TP / (TP + FN)
print("Precision:", tp / (tp + fp))  # Precision = TP / (TP + FP)
print("Recall:", tp / (tp + fn))  # Recall = TP / (TP + FN)
print("F1 Score:", 2 * (tp / (tp + fp)) * (tp / (tp + fn)) / ((tp / (tp + fp)) + (tp / (tp + fn))))  # F1 Score = 2 * (Precision * Recall) / (Precision + Recall)
print("Accuracy:", (tp + tn) / (tp + tn + fp + fn))
```
